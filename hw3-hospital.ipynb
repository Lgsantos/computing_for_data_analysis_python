```python
list(set(tmp['ownership'].values))
```

```python
from scipy import stats

fig, ax = plt.subplots(3,3)
plt.subplots_adjust(wspace = .1, hspace = .4)
own_list = list(set(tmp['ownership'].values))
for i in xrange(3):
    for j in xrange(3): 
        
        xs = ax[i, j]
        own = own_list[i + j*3]
        axtmp = tmp[tmp['ownership'] == own]
        slope, intercept, r_value, p_value, std_err = stats.linregress(axtmp['patients'],axtmp['rate'])
        x = np.linspace(0,1200,20)
        xs.plot(axtmp['patients'],axtmp['rate'],'o', alpha=0.5, markersize=5)
        xs.plot(x,slope*x+intercept)
        xs.set_title(own[:25])
        xs.set_xlim(0,1200)
        xs.set_ylim(10,22)

plt.show()
```

##Ranking hospitals by outcome in a state

```python
tmp = tmp_stack_re.reset_index()
print tmp.head()
```

```python
groups = tmp.groupby(['state','type'])
```

```python
def add_rank(g):
    g = g.sort('value')
    rank = range(len(g))
    g['rank'] = rank
    return g
```

```python
ranked = groups.apply(add_rank)
print ranked.head()
```

Example:

    > rankhospital("TX", "heart failure", 4)
    [1] "DETAR HOSPITAL NAVARRO"
    > rankhospital("MD", "heart attack", "worst")
    [1] "HARFORD MEMORIAL HOSPITAL"

```python
re_ranked = ranked[['name','value','rank']].reset_index()
re_ranked = re_ranked.set_index(['state','type','rank'])
```
